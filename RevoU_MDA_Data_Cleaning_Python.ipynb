## 3. Handling Missing Value
Terdapat dua pilihan dalam handling missing value, yaitu menghapus atau mengisikan suatu value

### a. Handling Missing Value - Mengisikan suatu value

```python
df.head(25)
```

```python
df['Country'] = df['Country'].fillna("USA")
df.head(25)
```

### b. Handling Missing Value - Menghapus missing value

```python
df.isnull().sum()
```

```python
df[ df.isnull().any(axis='columns') ]
```

```python
df.dropna(subset = ['InvoiceDate'], inplace=True)
```

```python
df.isnull().sum()
```

## 5. Handling Duplicated Data
Data duplicate akan dihapus hingga hanya menyisakan satu saja yang unique

```python
df = df.drop_duplicates()
```
